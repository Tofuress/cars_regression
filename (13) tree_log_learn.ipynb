# Decision Tree с логарифмом цены

````python
import pandas as pd
import numpy as np
````

````python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.tree import DecisionTreeRegressor
import joblib
import pickle
````

```
Загрузка обучающей и тестовой выборок с логарифмом цены и LabelEncoder
```

````python
X_train, X_test, y_train, y_test = joblib.load("data/data_split_label_log.pkl")
````

````python
model = DecisionTreeRegressor(random_state=42)
````

```
Обучение и сохранение модели
```

````python
model.fit(X_train, y_train)
joblib.dump(model, "models/decision_tree_log_model.pkl")
````

```
Предсказание на тестовой выборке
```

````python
y_pred_log = model.predict(X_test)
````

````python
y_pred = np.exp(y_pred_log)
````

```
Оценка модели
```

````python
tree_mse = mean_squared_error(y_test, y_pred)
tree_rmse = np.sqrt(tree_mse)
tree_mae = mean_absolute_error(y_test, y_pred)
tree_r2 = r2_score(y_test, y_pred)
````

```
Вывод результатов
```

````python
pd.options.display.float_format = '{:_.2f}'.format
pd.DataFrame({
    'Метод оценки': ['Среднеквадратическая ошибка (MSE)', 'Среднеквадратическая ошибка (RMSE)', 'Средняя абсолютная ошибка (MAE)', 'Коэффицент детерминации (R^2)'],
    'Результаты': [tree_mse, tree_rmse, tree_mae, tree_r2]
})
````

```
Сохранение метрик в отдельный файл
```

````python
metrics = {
    "model_name": "Decision Tree Log",
    "MSE": tree_mse,
    "RMSE": tree_rmse,
    "MAE": tree_mae,
    "R2": tree_r2
}
````

````python
with open("metrics/tree_log_metrics.pkl", "wb") as f:
    pickle.dump(metrics, f)
````
