```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#df = df[df["year"] >= 2018]
drp = [
    "Ссылка", "VIN", "Номер кузова",
    "Ошибка_ст", "Ошибка_знач", "Скрыто", "Проверено",
    "Кол-во просмотров", "Комплектация",
    "Особые отметки", "Поколение", "Рестайлинг",
    "Рефрижератор", "Спальник", "Объем кузова",
    "Категория ТС", "Тип техники", "Колёсная формула",
    "Количество осей", "Масса", "Грузоподъемность",
    "Длина кузова", "Тип кабины", "Рабочая ширина",
    "Моточасы", "Оборудование", "Высота седла",
    "Бренд ХОУ", "Объем ковша", "Длина стрелы",
    "Грузоподъемность стрелы", "Высота вышки",
    "Высота подъема", "Ходовая часть",
    "Владелец","Топливо", "Руль", "Регион", "Макро-регион", "Цвет", "Состояние", "Страна производства", "Количество мест", "Владельцы", "Пропуски в данных"
]

df = df.drop(columns=drp, errors="ignore")
```

```python
#df[df['Тип кузова'] == 'открытый кузов']
df["Дата размещения объявления"] = pd.to_datetime(df["Дата размещения объявления"], format='%Y-%m-%d')
df = df[df["Дата размещения объявления"].dt.year >= 2018]
#df["Дата размещения объявления"].min()
df.info()
```

```python
df[df['Тип кузова'] == 'автобус//бортовой грузовик']
```

```python
df.size
```

```python
valid_bodies = [
    "купе", "лифтбек", "открытый", "открытый кузов",
    "седан", "хэтчбек 3 дв.", "хэтчбек 5 дв.",
    "универсал", "минивэн", "джип/suv 5 дв.", "джип/suv 3 дв."
]

# Список слов для исключения (грузовые и спецтехника)
invalid_keywords = [
    "автобус", "грузовик", "бортовой", "шасси", "пикап", 
    "цельнометаллический", "фургон", "седельный тягач", "рефрижератор"
]

# Удаляем строки с //
df = df[~df["Тип кузова"].str.contains("//", na=False)]

# Удаляем строки с грузовыми ключевыми словами
mask_invalid = df["Тип кузова"].str.contains("|".join(invalid_keywords), case=False, na=False)
df = df[~mask_invalid]

# Оставляем только корректные легковые кузова
df = df[df["Тип кузова"].isin(valid_bodies)]

# Проверка
print(df["Тип кузова"].value_counts())
```

```python
df.size
```

```python
total_nan_count = df['Метка'].isnull().sum().sum()
total_nan_count
```

```python
req_col = list(df.columns)
df = df.dropna(subset=req_col)
df.size
```

```python
total_nan_count = df.isnull().sum().sum()
total_nan_count
```
